```python
import _mysql_connector
import mysql.connector
```

```python
db = mysql.connector.connect(
    host="localhost",
    user="root",
    password="riyash@1515",
    database="view1"
)

# Create a cursor object
cursor = db.cursor()

# Create table if it doesn't exist
cursor.execute("CREATE TABLE IF NOT EXISTS tasks (id INT AUTO_INCREMENT PRIMARY KEY, description VARCHAR(255), completed BOOLEAN)")
```

```python
# Function to add task to the database
def add_task(description):
    sql = "INSERT INTO tasks (description, completed) VALUES (%s, %s)"
    values = (description, False)
    cursor.execute(sql, values)
    db.commit()

# Function to get all tasks from the database
def get_tasks():
    sql = "SELECT * FROM tasks"
    cursor.execute(sql)
    return cursor.fetchall()

# Function to mark a task as completed
def mark_complete(task_id):
    sql = "UPDATE tasks SET completed = True WHERE id = %s"
    values = (task_id,)
    cursor.execute(sql, values)
    db.commit()

# Function to delete a task from the database
def delete_task(task_id):
    sql = "DELETE FROM tasks WHERE id = %s"
    values = (task_id,)
    cursor.execute(sql, values)
    db.commit()

# Main function to interact with the user
def main():
    while True:
        print("\n1. Add task")
        print("2. View tasks")
        print("3. Mark task as complete")
        print("4. Delete task")
        print("5. Quit")

        choice = input("Enter choice (1-5): ")

        if choice == "1":
            description = input("Enter task description: ")
            add_task(description)
            print("Task added successfully!")

        elif choice == "2":
            tasks = get_tasks()
            if len(tasks) == 0:
                print("No tasks found.")
            else:
                for task in tasks:
                    print(f"{task[0]}. {task[1]} - {'Completed' if task[2] else 'Incomplete'}")

        elif choice == "3":
            task_id = input("Enter task ID: ")
            mark_complete(task_id)
            print("Task marked as complete.")

        elif choice == "4":
            task_id = input("Enter task ID: ")
            delete_task(task_id)
            print("Task deleted.")

        elif choice == "5":
            break

        else:
            print("Invalid choice.")

    db.close()

if __name__ == "__main__":
    main()
```
